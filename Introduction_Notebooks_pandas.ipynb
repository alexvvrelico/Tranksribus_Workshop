## Understanding `JSON`, `GeoJSON`, Polygons & Basic GIS Concepts

Before we can plot people on a map, we need to understand how geographic shapes are represented in digital form. This is called GIS (Geographic Information Systems).

Since our historical sources often do not provide precise coordinates, we will reconstruct the Historical Neighbourhoods of Dudelange manually and store them in a simple format.

This leads us to three key concepts:

## 4.1. What is `JSON`?

**JSON** (JavaScript Object Notation) is a lightweight text format used everywhere to store and exchange data

It is **arborescent**, meaning that its structure is organized like a tree, creating a hierarchy of data.


Example:

```json
{
  "city": "Dudelange",
  "info": {
    "country": "Luxembourg",
    "population": 22184
  },
  "neighbourhoods": [
    {"name": "Quartier Italien"},
    {"name": "Brill"},
    {"name": "Gare-Usines"}
  ]
}
```

## 4.2. What is GeoJSON?

It is simply JSON extended for geographical data

In our workshop:

We will load a GeoJSON containing the Historical Neighbourhoods of Dudelange.
These polygons were drawn manually based on historical maps and local knowledge.

## 4.3. What is a Polygon in GIS?

A **polygon** is a closed shape defined by a sequence of coordinate pairs, used to represent:

* neighbourhood boundaries
* land parcels
* administrative units
* historical areas
* industrial sites

A **polygon** must start and end at the same coordinate.

In our case:

Each **historical neighbourhood** of Dudelange is represented as one polygon.
Each **resident** is then associated with the neighbourhood polygon they lived in.

## 4.4. What is a Shapefile ?

It is a traditional GIS file format (from the 1990s), widely used in GIS software (**QGIS**, **ArcGIS**).

## 4.5. What is an API in GIS?

An **API (Application Programming Interface)** is a service that lets you request data from a server.

In GIS, APIs are used to obtain:
* city boundaries
* administrative units
* roads and buildings
etc.

# Section 6: The `math` library, Shannon diversity and χ² test

For this, we will use the most common Python built-in library called `math`and we import `log`.

````python
# from math import log
````

The Shannon index measures diversity:

* higher value = many groups + more balanced proportions

* lower value = one group dominates

Here we compute diversity of origins within each neighbourhood.

As this is a complex move, we need to create our own **function**. This is where Python becomes interesting.

We begin a **loop** by using `def`, followed by the name of our function. Here: `shannon_entropy()`. `series` refers to the imput data the function will receive. In our case, it is a **Pandas series** containing the values of one variable.

The **keyword** `return` defines what the function **outputs**.
Here, a single number representing the **diversity** of the **series**.

````python
'''

def shannon_entropy(series):
    freqs = series.value_counts(normalize=True)
    return -(freqs * freqs.apply(lambda p: log(p))).sum()

diversite = (
    df.groupby("Quartier")["Orig_country"]
            .apply(shannon_entropy)
            .reset_index(name="diversite_shannon")
)

diversite.sort_values("diversite_shannon", ascending=False)
````
